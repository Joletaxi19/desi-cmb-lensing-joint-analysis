## A heuristic version of Zheng et al. 2007

```python
from scipy.special import erfc
# from sandy's paper
# https://desi.lbl.gov/DocDB/cgi-bin/private/RetrieveFile?docid=7374;filename=DESI_HOD_LRG%20%281%29.pdf;version=2
Mc = 10**12.89
M1 = 10**14.08
sigma = 0.27
alpha = 1.20
kappa = 0.65

Nbar_central = lambda M: erfc(np.log10(Mc/(0.68*M))/np.sqrt(2)/sigma)/2 # 0.68 accounts for M_sun -> M_sun/h
Nbar_sat = lambda M: (np.maximum((0.68*M)-kappa*Mc,0.)/M1)**alpha*Nbar_central(M)
```

```python
logmasses = np.linspace(28,34,50)

plt.plot(logmasses,Nbar_central(np.exp(logmasses)))
plt.plot(logmasses,Nbar_sat(np.exp(logmasses)))

plt.xlabel(r'log(mass)')
```

# Making (masked) maps and comparing power spectra to the true samples

```python
def get_map(isamp):
    pref  = "lrg_s{:02d}".format(isamp)
    db ='/pscratch/sd/m/mwhite/DESI/MaPar/maps/'
    lrg=hp.read_map(db+pref+'_del.hpx2048.fits')
    msk=hp.read_map(db+pref+'_msk.hpx2048.fits')
    lrg=hp.ma(lrg)
    lrg.mask=(msk<0.5)
    lrg -= np.mean(lrg)
    return lrg,msk

# true maps
lrg_s1,msk_s1 = get_map(1)
lrg_s2,msk_s2 = get_map(2)
lrg_s3,msk_s3 = get_map(3)
lrg_s4,msk_s4 = get_map(4)
```

```python
def make_map(catalog,dN_dz,nside=2048,lnM_min=20,lnM_max=80,mask=None,f=1):
    lnM =  np.log(catalog[:,3])
    I = np.where((lnM>lnM_min) & (lnM<lnM_max))
    
    # impose mass cut
    z =     catalog[:,0][I]
    theta = catalog[:,1][I]
    phi =   catalog[:,2][I]
    M =     catalog[:,3][I]
    
    # Apply Zheng et al. weighting
    N_central = 1.-np.ceil(np.random.random(size=len(z)) - Nbar_central(M))
    N_sat = np.random.poisson(lam=Nbar_sat(M))
    weights = N_central + N_sat
    
    # randomly downsample to increase shot noise
    I = np.random.choice(range(len(z)), int(f*len(z)), replace=False)
    z =     z[I]
    theta = theta[I]
    phi =   phi[I]
    weights = weights[I]
    
    # reweight to get dN/dz right
    # first "flatten" the dN/dz distribution
    nbar,zedges = np.histogram(z,bins=100)
    zcs = (zedges[:-1]+zedges[1:])/2.
    weights /= interp1d(zcs,nbar,bounds_error=False,fill_value=1e30)(z)
    # and then rescale to the target dN/dz
    weights *= interp1d(dN_dz[:,0],dN_dz[:,1],bounds_error=False,fill_value=0)(z)
    
    Map = np.zeros((hp.nside2npix(nside)))
    pix = hp.ang2pix(nside, theta, phi)
    np.add.at(Map, pix, weights)
    delta = Map/np.mean(Map)-1.
    lrg=hp.ma(delta) # mock lrg overdensity
    
    if mask is not None:
        lrg.mask=(mask<0.5)
        lrg -= np.mean(lrg)
        
    return lrg,z,weights

# mock maps
lnM_min = 20
delta_s1,z1,weights1 = make_map(catalog_s1,dN_dz_s1,mask=msk_s1,lnM_min=lnM_min,f=0.5)
delta_s2,z2,weights2 = make_map(catalog_s2,dN_dz_s2,mask=msk_s2,lnM_min=lnM_min,f=0.55)
delta_s3,z3,weights3 = make_map(catalog_s3,dN_dz_s3,mask=msk_s3,lnM_min=lnM_min,f=0.55)
delta_s4,z4,weights4 = make_map(catalog_s4,dN_dz_s4,mask=msk_s4,lnM_min=lnM_min,f=0.4)
```

```python
hp.mollview(delta_s1)
```

```python
nbins=50

plt.hist(z1,weights=weights1,density=True,bins=nbins,color='C0',alpha=0.8)
plt.hist(z2,weights=weights2,density=True,bins=nbins,color='C1',alpha=0.8)
plt.hist(z3,weights=weights3,density=True,bins=nbins,color='C2',alpha=0.8)
plt.hist(z4,weights=weights4,density=True,bins=nbins,color='C3',alpha=0.8)

plot_target_dist(dN_dz_s1,'cyan')
plot_target_dist(dN_dz_s2,'yellow')
plot_target_dist(dN_dz_s3,'lime')
plot_target_dist(dN_dz_s4,'lightcoral')

plt.xlabel(r'$z$')
plt.ylabel(r'$\propto\,dN/dz$')
plt.xlim(0.2,1.2)
plt.show()
```

```python
Cl1 = hp.anafast(lrg1,map2=tSZ,lmax=2000)
Cl2 = hp.anafast(lrg2,map2=tSZ,lmax=2000)
Cl3 = hp.anafast(lrg3,map2=tSZ,lmax=2000)
Cl4 = hp.anafast(lrg4,map2=tSZ,lmax=2000)
```
